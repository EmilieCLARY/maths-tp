## $\nabla$ - Méthode du gradient

```python
# Définition de la fonction gradiant

def gradiant(init_point, step, N, delta, epsilon, function):
    
    # Rappel des paramètres d'entrée de fonction
    show("Méthode du gradient sur la fonction f",function) 
    
    pn = [] # Liste des points de la suite
    
    pn.append(init_point)
    function_gradiant = f.gradient() # Calcul des dérivées de la fonction f
    pn.append((init_point[0] + (step * function_gradiant(init_point[0],init_point[1]))[0],init_point[1] + (step * function_gradiant(init_point[0],init_point[1]))[1]))
    i = 0 # Initialisation de l'itérateur à zéro
    cond = False # Initialisation de notre condition à faux (afin de rentrer au moins dans la boucle)
    
    while (i != N or cond == True):
        pnx = pn[len(pn)-1][0] + (step * function_gradiant(pn[len(pn)-1][0],pn[len(pn)-1][1]))[0] # On calcule la composante en x de P_{n+1} ...
        pny = pn[len(pn)-1][1] + (step * function_gradiant(pn[len(pn)-1][0],pn[len(pn)-1][1]))[1] # ... On calcule la composante en y de P_{n+1} ...
        pn.append((n(pnx),n(pny)))  # ... Et on ajoute leur approximation numérique à la liste de point sous forme de tuple !
        cond = (vector([pn[len(pn)-1][0],pn[len(pn)-1][1]]).norm()  - vector([pn[len(pn)-2][0],pn[len(pn)-2][1]]).norm()) < delta and abs(function(pn[len(pn)-1][0],pn[len(pn)-1][1]) - function(pn[len(pn)-2][0],pn[len(pn)-2][1])) < epsilon
        i += 1

    return pn

    
```

On remarque bien que la fonction $f$ possède (dans l'intervalle $x\in[-2,2]$ et $y\in[-2,2]$) trois minimums locaux clairement défini

```python
# Pensez à bien relancer le bloc au dessus afin de modifier celui-ci

grad = gradiant((1,2),0.1,50,0.001,0.001,f)
graph = contour_plot(f,(x,-2,3),(y,-2,2),fill = False, cmap = True)
graph += plot_vector_field((f.gradient()),(x,-2,3),(y,-2,2), color = "red")

for dot in grad:
    graph += point(dot, color = "blue")    
    
show("Graphique de la fonction f",f)
show(graph)
```

On remarque qu'avec la méthode $\verb|gradient()|$ codée ci-dessus, le maximum que la fonction nous renvoie est bien un maximum local et non un maximum global.

## $\mathcal{H}$ - Méthode de Newton

```python
# Définition de la méthode de Newton

def newton(init_point, step, N, delta, epsilon, function):
    
    # Rappel des paramètres d'entrée de fonction
    show("Méthode du gradient sur la fonction f",function) 
    
    pn = [] # Liste des points de la suite
    
    pn.append(init_point)
    function_derivative = function.derivative()
    pn.append((pn[len(pn)-1][0] - (function(pn[len(pn)-1][0],pn[len(pn)-1][1]))/function_derivative(pn[len(pn)-1][0],pn[len(pn)-1][1])[0], pn[len(pn)-1][1] - (function(pn[len(pn)-1][0],pn[len(pn)-1][1]))/function_derivative(pn[len(pn)-1][0],pn[len(pn)-1][1])[1]))
    show(pn[len(pn)-1])
    i = 0 # Initialisation de l'itérateur à zéro
    cond = False # Initialisation de notre condition à faux (afin de rentrer au moins dans la boucle)
    
    while (i != N or cond == True):
        pnx = pn[len(pn)-1][0] - (function(pn[len(pn)-1][0],pn[len(pn)-1][1]))/function_derivative(pn[len(pn)-1][0],pn[len(pn)-1][1])[0] # On calcule la composante en x de P_{n+1} ...
        pny = pn[len(pn)-1][1] - (function(pn[len(pn)-1][0],pn[len(pn)-1][1]))/function_derivative(pn[len(pn)-1][0],pn[len(pn)-1][1])[1] # ... On calcule la composante en y de P_{n+1} ...
        pn.append((n(pnx),n(pny)))  # ... Et on ajoute leur approximation numérique à la liste de point sous forme de tuple !
        

        cond = (vector([pn[len(pn)-1][0],pn[len(pn)-1][1]]).norm()  - vector([pn[len(pn)-2][0],pn[len(pn)-2][1]]).norm()) < delta and abs(function(pn[len(pn)-1][0],pn[len(pn)-1][1]) - function(pn[len(pn)-2][0],pn[len(pn)-2][1])) < epsilon
        i += 1

    return pn

    
```
