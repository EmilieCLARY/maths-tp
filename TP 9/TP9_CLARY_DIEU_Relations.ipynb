## Exercice 1 :


Combien y a-t-il de relations binaires distinctes sur En ?

Pour confirmer votre réponse, générer toutes les relations possibles pour n = 3, sous deux formes :
    
    • Liste des couples d’éléments en relation ;
    • Matrice d’adjacence.

```sage
def parties(E) :

    while(len(E) > 0): # On fait une copie de l'ensemble
        partie = copy(E)
        partie.pop()

        list_ = copy(parties(partie))

        listeR = []

        for i in list_:
            init = copy(i)
            listeR.append(init)
            nE = copy(i)
            nE.append(E[-1])
            listeR.append(nE)

        return listeR
    return [[]]


# On pose notre ensemble :
En = [(x,y) for x in range(3) for y in range(3)]
```

```sage
def mat_adj(parties, size):
    listeRetour = []
    for relation in parties:
        matrice = matrix(size)
        for element in relation:
            matrice[element] = 1
        listeRetour.append(matrice)
    return listeRetour

allRelations = mat_adj(parties(En), 3)

len(allRelations)
```

```sage
def cap(A) :

    _size = A.nrows
    
    for n in range(_size) :
        for j in range(_size) :
            if(A[i, j] > 1) :
                A[i, j] = 1

    return A

def oplus(A, B) :
    return cap(A+B)

def otimes(A,B) :
    return cap(A*B)

def is_contained(A,B) :
    return oplus(A,B) == B
    
```

```sage
def is_reflexive(matrice, size):

    for x in range(size) :

        # On teste les valeurs de la diagonale.
        if(matrice[x,x] != 1)   :
            return False          
    
    return True
```

```sage
def is_symetric(matrice, size):

    for x in range(size):
        for y in range(size):
            
            if x < y:
                # On teste la valeur de la position actuelle et inverse par rapport à la diagonale.
                if matrice[x,y] != matrice[y,x]:
                    return False
    
    return True
```

```sage
def is_transitive(matrice):
    n=matrice.nrows()
    for j in range(n):
        for i in range(n):
            if(matrice[i][j]== 1):
                for k in range(n):
                    if(matrice[j][k] == 1 and matrice[i][k]!=1):
                        return false
    return true
```

```sage
def is_antisymmetric(matrice, size) :

    for x in range(size):
        for y in range(size):
            
            if x < y:
                # Deux 1 ne peuvent pas êtresymétriques par rapport à la diagonale.
                if (matrice[x,y] == 1 and matrice[y,x] == 1) :    
                    return False

    return True
```

Nous commençons par nous convaincre que c'est bien une relation d'équivalence sur l'ensemble des relations.

Relation d'équivalence = Réflexive, Transitive et Symétrique.

- Or T est une relation réflexive car si R est en relation T avec R, alors R est inclus dans T. Ce qui équivaut à R inclus dans T.

- Ensuite, T est une relation transitive car soit R, S, U, si R est en relation T avec S et S est en relation T avec U alors R est inclus dans T ce qui équivaut à S inclus dans T et S inclus dans T équivaut à U inclus dans T. Au final on a R inclus dans T équivalent à U inclus dans T.

- Pour finir, T est une relation symétrique car si R est en relation T avec S alors R est inclus dans T ce qui équivaut à S inclus dans T. Inversement, on a bien S en relation T avec R qui donne S inclus dans T équivalent à R inclus dans T.

On a bien une relation d'équivalence.
