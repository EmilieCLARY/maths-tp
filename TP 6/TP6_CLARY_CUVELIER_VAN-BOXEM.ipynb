## Exercice 1
- Visualiser, à l’aide de complex_plot, les fonctions complexes suivantes et les commenter. (position et multiplicité des zéros et des pôles, tendances asymptotiques . . . )

$$ f(z) = z^2 - z - 1 $$

- Visualisation graphique :

```python
z = var('z')
t = var('t')
f(z) = z^2 - z - 1

complex_plot(f, (-1,2), (-2,2))  + plot(lambda t: arg(f(t+i)), -1, 2)
```

On voit que la fonction possède deux 0 : $ (-0.6; 0)$ et $(1.6; 0)$.

Autour de chaque 0 (les points noirs correspondant aux coordonnées).
On peut en déduire qu'il s'agit de deux racines.

En $x = 0.5$, on a une ligne blanche qui correspond à un changement brutal de l'argument (négatif à positif).

$$ g(z) = \frac{1}{z} + z $$

- Visualisation graphique :

```python
z = var('z')
t = var('t')
g(z) = (1/z) + z

complex_plot(g, (-3/2,3/2), (-3/2,3/2)) +  plot(lambda t: arg(g(t+i)), -1, 2)
```

On voit que la fonction possède deux 0 : $ (0; 1)$ et $ (0; -1)$.

Également, on peut en déduire qu'il s'agit de deux racines.

Il y a aussi un point blanc en $ (0 ; 0)$, cela correspond à la valeur interdite.

Finalement, on voit que l'argument effectue un saut en 0 tout en restant positif.

$$  h(z) = \cos(z) $$
- Visualisation graphique :

```python
z = var('z')
h(z) = cos(z)

complex_plot(h, (-7,7), (-7,7))
```

Puis on zoom sur un intervalle :

```python
t = var('t')
z = var('z')
h(z) = cos(z)

complex_plot(h, (-pi,pi), (-pi,pi)) +  plot(lambda t: arg(h(t+i)), -5, 5)
```

On voit que la fonction possède deux 0 : $(-\frac{\pi}{2} ; 0)$ et $(\frac{\pi}{2} ; 0)$.

Il n'y a pas de point blanc, donc aucune valeur interdite.

On voit ici que l'argument est périodique d'un saut de $\pi$

## Exercice 2
- Cela peut nous aider à comprendre la façon dont les fonctions analytiques sont représentées par leurs
développements en séries de puissances. Observez les sommes, et quelques sommes partielles, des séries suivantes.
Arrivez-vous à déterminer les rayons de convergence graphiquement ?

$$ \frac{z^2 + z + 1}{2 - z} = \frac{1}{2} + \frac{3}{4} z + \frac{7}{2} \sum_{n=2}^{\infty} \frac{z^n}{2^n} $$

On pose : $$ f(z) = \frac{z^2 + z + 1}{2 - z} $$ 
Puis on observe sa représentation graphique :

```python
var('z')
f(z) = (z^2 + z + 1)/(2 - z)

complex_plot(f, (-3,3), (-3,3))
```

En ajoutant la série des sommes partielles, on a : 

```python
# Définition de la fonction Somme Partielle de F
def sfz(z,n):
    somme = (1/2)+(3/4)*z+(7/2)
    for i in range (n):
        somme += (z^i)/2^i
    return somme

complex_plot(sfz(z,100), (-3,3), (-3,3), aspect_ratio = 1)
    
```

On remarque que le rayon de convergence est égal à $2$.

$$ \frac{1}{z} = \sum_{n=0}^{\infty} \frac{(-1)^n}{3^{n+1}} (z - 3)^n $$

On pose : $$ g(z) = \frac{1}{z} $$ 
Puis on observe sa représentation graphique :

```python
z = var('z')
g(z) = (1/z)

complex_plot(g,(-2,7),(-4,5))
```

Rajoutant la série des sommes partielles.

```python
def gsz(z,n):
    somme = 0
    for i in range (n):
        somme += (((-1)^i)/(3^(i+1)))*(z-3)^i
        
    return somme

complex_plot(gsz(z,100),(-2,7),(-4,5), aspect_ratio = 1)
```

Le rayon de convergence est égal à $3$.

$$ \arctan(z) = \sum_{n=0}^{\infty} \frac{(-1)^n}{2n + 1} z^{2n+1} $$

```python
z = var('z')
h(z) = arctan(z)

complex_plot(h, (-2,2), (-2,2))
```

```python
def shz(z,n):
    somme = 0
    for i in range (n):
        somme += (((-1)^i)/(2*i+1))*z^(2*i+1)
    return somme

complex_plot(shz(z,100), (-2,2), (-2,2), aspect_ratio = 1)
```

Le rayon de convergence est de $1$.

## Exercice 3

- Reprendre la visualisation des fonctions précédentes en les post-composant par q. Qu’observez-vous ? (Portez
attention, notamment, à la façon dont les courbes noires se croisent.)
Vous êtes bien sûr encouragés à expérimenter avec d’autres fonctions / images dans le plan d’arrivée.

```python
# On defini la fonction q qui remplace la fonction identité 
z = var('z')
q(z) = sin(pi*z.real()) * sin(pi*z.imag())
complex_plot(q, (-2, 2), (-2, 2))
```

```python
f(z) = z^2 - z - 1
t = var('t')
show(complex_plot(f, (-1, 2), (-2, 2)) + plot(lambda t: arg(f(t+i)), -1, 2))
show(complex_plot(q(f), (-1, 2), (-2, 2)))
print("Par à la fonction Q(), on visualise deux axes de symétries : l'axe des réels et l'axe des imaginaire passant par x = 0.5. On avait vu que les points de coordonnées (0.5;a), a appartient aux entiers, tendent vers l'infini car on a un saut de l'argument (du négatif vers le positif). La présence de cette droite explique l'axe de symétrie. De ces axes de symétries, on en deduit un centre de symétrie en leur intersection (0.5;0). Ces axes sont caracterisés par des points nuls.")
```

```python
g(z) = (1/z) + z
show(complex_plot(g, (-3/2,3/2), (-3/2,3/2)) + plot(lambda t: arg(g(t+i)), -1.5, 1.5))
show(complex_plot(q(g), (-3/2,3/2), (-3/2,3/2)) + plot(lambda t: arg(q(g)(t+i)), -1.5, 1.5))
print("On observe que les pôles ne sont pas bien visible (confondus avec le restes des points nuls), mais on voit bien la valeur interdite (quand x = (0;0)).")
```

```python
h(z) = cos(z)
show(complex_plot(h, (-pi, pi), (-pi, pi)) + plot(lambda t: arg(h(t+i)), -5, 5))
show(complex_plot(q(h), (-pi, pi), (-pi, pi)))
print("On a une periodicité de l'arguement de période pi. L'argument s'annule tous les Pi. Des droites verticales passent par les points (r, 0) où r est une racine de h. Ces droites sont des axes de symétries.")
```
